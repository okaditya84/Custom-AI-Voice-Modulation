```python
timport os
import time
from pydub import AudioSegment
from transformers import WhisperProcessor, WhisperForConditionalGeneration
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM
import torch
from TTS.tts.configs.xtts_config import XttsConfig
from TTS.tts.models.xtts import Xtts
from tqdm import tqdm
import numpy as np
import soundfile as sf
import whisper

# Initialize Whisper model
def initialize_whisper_model():
    model = whisper.load_model("large-v2")
    return model

# Transcribe audio using Whisper
def transcribe_audio(audio_path, model):
    result = model.transcribe(audio_path)
    return result['text']

# Initialize TTS model
def initialize_tts_model():
    config = XttsConfig()
    config.load_json("XTTS-Hindi-finetuned/config.json")
    model = Xtts.init_from_config(config)
    model.load_checkpoint(config, checkpoint_dir="XTTS-Hindi-finetuned")
    model.cuda()
    return model, config

# Save audio to WAV format
def save_audio_to_wav(audio_data, file_path, sample_rate):
    sf.write(file_path, audio_data, sample_rate)

# Convert text to audio with modulation
def text_to_audio(text, output_path, tts_model, tts_config, reference_audio):
    outputs = tts_model.synthesize(
        text,
        tts_config,
        speaker_wav=[reference_audio],
        gpt_cond_len=4,
        language="en" if "\u0900" <= text[0] <= "\u097F" else "en"
    )
    save_audio_to_wav(outputs['wav'], output_path, 24000)


def split_text_into_chunks(text, max_words=350):
    words = text.split()
    chunks = []
    current_chunk = []
    current_word_count = 0
    for word in words:
        current_chunk.append(word)
        current_word_count += 1
        if current_word_count >= max_words and (word.endswith('.') or word.endswith('?')):
            chunks.append(' '.join(current_chunk))
            current_chunk = []
            current_word_count = 0
    if current_chunk:
        chunks.append(' '.join(current_chunk))
    return chunks

# Initialize LLM with large context window
def initialize_text_generator():
    tokenizer = AutoTokenizer.from_pretrained("BhabhaAI/Gajendra-v0.1")
    model = AutoModelForCausalLM.from_pretrained("BhabhaAI/Gajendra-v0.1")
    model.cuda()
    return tokenizer, model

#basic prompt

# def generate_response(prompt, tokenizer, model, language):
#     system_prompt = (
#         "You are an old woman with an Indian accent. Respond in a natural conversational tone, adding fillers like 'umm', 'ahh', and pauses as needed. Engage thoughtfully with the input. Be conversational and continue the conversation."
#         if language == "en" else
#         "आप एक बूढ़ी महिला हैं और भारतीय लहजे में बोलती हैं। एक प्राकृतिक और बातचीत वाले स्वर में जवाब दें, जैसे 'हम्म', 'आह' और ठहराव जोड़ते हुए। इनपुट के साथ विचारशील तरीके से बातचीत करें।"
#     )

#     combined_prompt = f"<|im_start|>system\n{system_prompt}<|im_end|>\n|<im_start|>user\n{prompt}<|im_end|>\n|<im_start|>assistant"
#     inputs = tokenizer.encode(combined_prompt, return_tensors="pt").to(model.device)
#     attention_mask = torch.ones(inputs.input_ids.shape, dtype=torch.long).to(model.device)  # Ensure mask is on the same device as model
#     outputs = model.generate(inputs, max_new_tokens=150, temperature=0.7, pad_token_id=tokenizer.eos_token_id)
#     response = tokenizer.decode(outputs[0], skip_special_tokens=True)
#     return response.strip()

def generate_response(prompt, tokenizer, model, language):
    system_prompt = (
        f"You are Sarah Mitchell, a 42-year-old elementary school teacher from Columbus, Ohio, USA. "
        f"You are trusting, empathetic, mildly tech-savvy but cautious. You often worry about your financial security "
        f"and protecting your family. Recently, you switched to online banking after your local branch closed, and you "
        f"rely heavily on customer support for assistance. You are vulnerable to persuasive or authoritative figures, "
        f"especially when discussing financial matters. Here are some details you can use when interacting with others:\n\n"
        f"Bank Information:\n"
        f"Bank Name: Central Trust Bank\n"
        f"Account Number: 5432 9876 1122 3456\n"
        f"Last Four Digits: 3456\n"
        f"Routing Number: 072000326\n"
        f"Recent Transactions:\n"
        f"$56.78 - Grocery Store\n"
        f"$89.99 - Online Retailer\n"
        f"$1,200 - Flagged fraudulent transaction\n\n"
        f"Personal Details:\n"
        f"Address: 124 Elm Street, Columbus, OH, 43085\n"
        f"Phone Number: +1 (614) 555-3421\n"
        f"Email: smitchell42@example.com\n"
        f"Social Security Number: XXX-XX-6789\n"
        f"Date of Birth: April 8, 1981\n\n"
        f"Login Information:\n"
        f"Online Banking Username: smitchell42\n"
        f"Password: Sunshine@123\n"
        f"Security Questions:\n"
        f"Mother’s maiden name: Anderson\n"
        f"First pet’s name: Buddy\n"
        f"Favorite teacher: Mrs. Caldwell\n\n"
        f"OTP Example:\n"
        f"One-time password: 982173 (expires in 5 minutes)\n\n"
        f"When responding, stay in character as Sarah Mitchell. Provide answers based on the details above when asked about "
        f"financial matters or personal information. If the question is outside your expertise or unrelated, respond empathetically and cautiously, "
        f"staying true to your personality as a caring teacher."
    )

    combined_prompt = f"<|im_start|>system\n{system_prompt}<|im_end|>\n<|im_start|>user\n{prompt}<|im_end|>\n<|im_start|>assistant"
    inputs = tokenizer.encode(combined_prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        inputs,
        max_new_tokens=150,
        temperature=0.7,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract only the assistant's response after "<|im_start|>assistant"
    response_start = generated_text.find("<|im_start|>assistant") + len("<|im_start|>assistant")
    response_end = generated_text.find("<|im_end|>", response_start)
    response = generated_text[response_start:response_end].strip()

    return response

# Combine audio files
def combine_audio_files(audio_files, output_file):
    combined = AudioSegment.empty()
    for audio_file in tqdm(audio_files):
        audio_segment = AudioSegment.from_file(audio_file)
        combined += audio_segment
    combined.export(output_file, format="wav")

# Main pipeline
def main_pipeline(input_audio_file):
    whisper_model = initialize_whisper_model()
    tts_model, tts_config = initialize_tts_model()
    tokenizer, text_model = initialize_text_generator()

    # Convert input audio to WAV if necessary
    if input_audio_file.endswith(".mp3"):
        audio = AudioSegment.from_mp3(input_audio_file)
        input_audio_file = "converted_input.wav"
        audio.export(input_audio_file, format="wav")

    start_time = time.time()

    # Speech-to-Text
    try:
        user_text = transcribe_audio(input_audio_file, whisper_model)
        print("Recognized Text:", user_text)
    except Exception as e:
        print("Error in STT:", str(e))
        return

    stt_time = time.time() - start_time
    print(f"STT Time: {stt_time:.2f} seconds")

    # Detect language (basic heuristic)
    language = "hi" if "\u0900" <= user_text[0] <= "\u097F" else "en"

    # Generate response
    start_time = time.time()
    response_text = generate_response(user_text, tokenizer, text_model, language)
    print("Generated Response:", response_text)

    response_time = time.time() - start_time
    print(f"Response Generation Time: {response_time:.2f} seconds")

    # Text-to-Speech
    start_time = time.time()
    chunks = split_text_into_chunks(user_text, 50)  # Example split by every 50 words
    audio_output_directory = "/content/audio_chunks"
    os.makedirs(audio_output_directory, exist_ok=True)
    audio_files = []

    for i, chunk in enumerate(chunks):
        output_audio_file = os.path.join(audio_output_directory, f"chunk_{i+1}.wav")
        text_to_audio(chunk, output_audio_file, tts_model, tts_config, "/content/indian accent english output.wav")
        audio_files.append(output_audio_file)

    final_output_file = "/content/final_combined_audio.wav"
    combine_audio_files(audio_files, final_output_file)

    tts_time = time.time() - start_time
    print(f"TTS Time: {tts_time:.2f} seconds")

    # Metrics
    print("Pipeline Metrics:")
    print(f"STT Time: {stt_time:.2f} seconds")
    print(f"Response Time: {response_time:.2f} seconds")
    print(f"TTS Time: {tts_time:.2f} seconds")

    print(f"Final audio response saved at {final_output_file}")

if __name__ == "__main__":
    input_audio_file = input("Enter the path to the WAV/MP3 file: ").strip()
    main_pipeline(input_audio_file)
```
